```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_excel("Superstore_USA.xlsx")
```

```python
dataset.head(5)
```

```python
dataset.shape
```

```python
dataset.isnull().sum()
```

```python
dataset["Product Base Margin"] = dataset["Product Base Margin"].fillna(dataset["Product Base Margin"].mean())
```

```python
dataset.isnull().sum()
```

```python
dataset["Order Priority"].value_counts()
```

```python
dataset["Order Priority"].unique()
```

```python
sns.countplot(x="Order Priority", data= dataset)
plt.title("Count of Order Priority")
plt.savefig("Count of Order Priority.jpg")
plt.show()
```

```python
sns.countplot(x="Order Priority", data= dataset)  
plt.title("Count of Order Priority")
plt.savefig("Count_of_Order_Priority.jpg") 
plt.show()
```

```python
x = dataset["Ship Mode"].value_counts().index
y = dataset["Ship Mode"].value_counts().values
```

```python
sns.countplot( x= "Ship Mode",data = dataset, hue = "Product Category")
plt.title("Ship Mode")
plt.savefig("Ship Mode.jpg") 
plt.show()
```
